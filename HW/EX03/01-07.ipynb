# 6.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

img_location = 'ZhongXinaSmaller.jpg'

img = cv2.imread(img_location)
h, w, c = img.shape

print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print('----------------------------------------------------')

img_pixels = img.reshape((-1, 3))

k = 16
kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(img_pixels)  # Set n_init explicitly to 10
labels = kmeans.predict(img_pixels)

avg_colors = []
for cluster_num in range(k):
    cluster_pixel_values = img_pixels[labels == cluster_num]
    avg_color = cluster_pixel_values.mean(axis=0)
    avg_colors.append(avg_color)

quantized_pixels = np.array([avg_colors[label] for label in labels])

quantized_image = quantized_pixels.reshape(img.shape).astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(quantized_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Improved Quantized Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()

cv2.imwrite('Improved_Quantized_Image.bmp', quantized_image)
print('Improved Quantized Image Saved')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_location = 'ZhongXinaSmaller.jpg'

img = cv2.imread(img_location)

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print('----------------------------------------------------')


def median_cut_quantization(img_pixels, num_colors):
    initial_box = list(range(img_pixels.shape[0]))
    boxes = [initial_box]
    
    while len(boxes) < num_colors:
        largest_range = -1
        box_to_split = None
        for box in boxes:
            for dim in range(3): 
                range_dim = img_pixels[box, dim].max() - img_pixels[box, dim].min()
                if range_dim > largest_range:
                    largest_range = range_dim
                    box_to_split = box
                    split_dim = dim
        
        box_sorted = sorted(box_to_split, key=lambda i: img_pixels[i, split_dim])
        median_idx = len(box_sorted) // 2
        left_box = box_sorted[:median_idx]
        right_box = box_sorted[median_idx:]
        boxes.remove(box_to_split)
        boxes.append(left_box)
        boxes.append(right_box)
    
    avg_colors = []
    for box in boxes:
        avg_color = img_pixels[box].mean(axis=0)
        avg_colors.append(avg_color)
    
    quantized_pixels = np.zeros_like(img_pixels)
    for box, avg_color in zip(boxes, avg_colors):
        for idx in box:
            quantized_pixels[idx] = avg_color
    
    return quantized_pixels

sample_image = cv2.imread(img_location)
sample_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)
img_pixels = sample_image.reshape((-1, 3))

k = 16 
quantized_pixels_median_cut = median_cut_quantization(img_pixels, k)
quantized_image_median_cut = quantized_pixels_median_cut.reshape(sample_image.shape).astype(np.uint8)


fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(sample_image)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(quantized_image_median_cut)
axes[1].set_title("Quantized Image (Median-Cut)")
axes[1].axis('off')

plt.tight_layout()
plt.show()

cv2.imwrite('Iquantized_Median_Cut_Image.bmp', cv2.cvtColor(quantized_image_median_cut, cv2.COLOR_BGR2RGB))
print('Iquantized Median Cut Image Saved')
```
