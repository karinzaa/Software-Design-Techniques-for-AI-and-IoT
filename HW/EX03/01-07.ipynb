```python
import cv2
import argparse
import matplotlib.pyplot as plt


img_location = 'ZhongXinaSmaller.jpg'
print('Image File Location')
print(img_location)


img = cv2.imread(img_location)
print('File loaded')

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print('----------------------------------------------------')


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Invert_img = 255 - gray_img


fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Input Image")
axes[0].axis('off')

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("Gray Image")
axes[1].axis('off')

axes[2].imshow(Invert_img, cmap='gray')
axes[2].set_title("Inverted Image")
axes[2].axis('off')

plt.tight_layout()
plt.show()


cv2.imwrite('Invert_img.bmp', Invert_img)
print('File Saved')


```

```python
import cv2
import argparse
import matplotlib.pyplot as plt


img_location = 'ZhongXinaSmaller.jpg'
print('Image File Location')
print(img_location)


img = cv2.imread(img_location)
print('File loaded')

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print('----------------------------------------------------')


new_width = int(w / 2)
new_height = int(h * (new_width / w))


resized_image = cv2.resize(img, (new_width, new_height))


fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Input Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Resized Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()


cv2.imwrite('Resized_Input_Image.bmp', resized_image)
print('Resized Image Saved')
```

```python
import cv2
import argparse
import matplotlib.pyplot as plt


img_location = 'ZhongXinaSmaller.jpg'
print('Image File Location')
print(img_location)


img = cv2.imread(img_location)
print('File loaded')

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print('----------------------------------------------------')


enlarged_width = int(w * 1.5)
enlarged_height = int(h * (enlarged_width / w))

enlarged_image = cv2.resize(img, (enlarged_width, enlarged_height))

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Input Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(enlarged_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Enlarged Image")
axes[1].axis('off')

h_enlarged, w_enlarged, c_enlarged = enlarged_image.shape
print('----------------------------------------------------')
print("Enlarged Image Information")
print("Dimension w x h = %s x %s" % (w_enlarged, h_enlarged))
print('----------------------------------------------------')

plt.tight_layout()
plt.show()

cv2.imwrite('Enlarged_Input_Image.bmp', enlarged_image)
print('Enlarged Image Saved')
```
