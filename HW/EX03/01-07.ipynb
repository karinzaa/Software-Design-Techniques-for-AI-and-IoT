```python
import cv2
import argparse
import matplotlib.pyplot as plt


img_location = 'https://github.com/karinzaa/Software-Design-Techniques-for-AI-and-IoT/blob/main/HW/EX03/ZhongXinaSmaller.jpg?raw=true'
parser = argparse.ArgumentParser(description='input image name')
parser.add_argument('--file', help="location of the image file")

# Set your image location here
img_location = 'ZhongXinaSmaller.jpg'
print('Image File Location')
print(img_location)


img = cv2.imread(img_location)
print('File loaded')

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print("Channels = %s" % (c))
print("Total Pixels of input image = %s pixels with %s channels" % (w*h, c))
print('----------------------------------------------------')


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Invert_img = 255 - gray_img


fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Input Image")
axes[0].axis('off')

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("Gray Image")
axes[1].axis('off')

axes[2].imshow(Invert_img, cmap='gray')
axes[2].set_title("Inverted Image")
axes[2].axis('off')

plt.tight_layout()
plt.show()


cv2.imwrite('Invert_img.bmp', Invert_img)
print('File Saved')


```

```python
import cv2
import argparse
import matplotlib.pyplot as plt


img_location = 'https://github.com/karinzaa/Software-Design-Techniques-for-AI-and-IoT/blob/main/HW/EX03/ZhongXinaSmaller.jpg?raw=true'
parser = argparse.ArgumentParser(description='input image name')
parser.add_argument('--file', help="location of the image file")

# Set your image location here
img_location = 'ZhongXinaSmaller.jpg'
print('Image File Location')
print(img_location)


img = cv2.imread(img_location)
print('File loaded')

h, w, c = img.shape
print('----------------------------------------------------')
print("Image Information")
print("Dimension w x h = %s x %s" % (w, h))
print("Channels = %s" % (c))
print("Total Pixels of input image = %s pixels with %s channels" % (w*h, c))
print('----------------------------------------------------')


# Rotating the input image 90 degrees clockwise
rotated_image = cv2.transpose(img)
rotated_image = cv2.flip(rotated_image, 1)

# Displaying the original and rotated images side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Input Image")
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Rotated Input Image")
axes[1].axis('off')

plt.tight_layout()
plt.show()

# Saving the rotated image
cv2.imwrite('Rotated_Input_Image.bmp', rotated_image)
print('Rotated Image Saved')
```
